```python
import pandas as pd

df = pd.read_excel("./MergedJualBeli2024_Excel.xlsx")
df.info()
```

```python
df.columns
```

```python
df.drop(columns=['Unnamed: 27','Helper: Check if Kode #_Jual !== Kode #_Beli','Helper: Recount Laba Jual', 'Helper: Cek Laba Asli dan Laba Recount'], axis=1, inplace= True)
df.columns
```

```python
numerical_df = df.select_dtypes(include=['int64','float64'])
numerical_df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
numerical_df['Kategori Barang Encoded'] = LabelEncoder().fit_transform(df["Nama Kategori Barang Barang & Jasa_Jual"])
numerical_df.columns
numerical_df
```

```python
coor_matrix = numerical_df.corr()
coor_matrix
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize with a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(coor_matrix, annot=True, cmap='coolwarm', center=0)
plt.title("Correlation Matrix")
plt.show()
```

```python
# Get correlations for 'feature_A'
target_feature = 'Laba_Jual'
correlations = numerical_df.corr()[target_feature].drop(target_feature)  # Exclude self-correlation

# Plot
plt.figure(figsize=(8, 6))
sns.barplot(x=correlations.index, y=correlations.values)
plt.xticks(rotation=90)
plt.title(f"Correlation of '{target_feature}' with Other Features")
plt.ylabel("Correlation Coefficient")
plt.axhline(0, color='black', linestyle='--')  # Add reference line at 0
plt.show()
```
