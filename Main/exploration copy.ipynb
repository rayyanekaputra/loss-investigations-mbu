```python
import pandas as pd

df = pd.read_excel("./MergedJualBeli2024_Excel.xlsx")
df.info()
```

```python
df.columns
```

```python
df.drop(columns=['Unnamed: 27','Helper: Check if Kode #_Jual !== Kode #_Beli','Helper: Recount Laba Jual', 'Helper: Cek Laba Asli dan Laba Recount'], axis=1, inplace= True)
df.columns
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your data is in a DataFrame called 'df'
# First, let's count transactions per month
transactions_per_month = df['Bulan_Jual'].value_counts().sort_index()

# Plotting
plt.figure(figsize=(12, 6))
transactions_per_month.plot(kind='bar', color='skyblue')
plt.title('Number of Transactions per Month')
plt.xlabel('Month')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
transactions_per_month.plot(kind='line', marker='o', color='red')
plt.title('Transaction Trend Over Months')
plt.xlabel('Month')
plt.ylabel('Number of Transactions')
plt.grid(True)
plt.show()
```
